```python
# Programming Final Exam Project
# World Cup Football Tournament System
# Group Number: 10
# Members: Athalla N.R. Anggorodi, Azhar Chukusheva, Muhammad Habibie Abdillah,
# Seongeun Park, Shanise Ruvimbo Mukwashi, Shokhruz Kabilov, Tarek Husain

import itertools

### Defines classes for teams, groups, and each stage
class team:
    def __init__(self, country, initials):
        self.country = country
        self.initials = initials
        self.win = 0
        self.loss = 0
        self.draw = 0
    def upd_res(self, result):    #function to update win/draw/loss results in group stage
        if result == "win":
            self.win += 1
        elif result == "loss":
                self.loss += 1
        else:
            self.draw += 0.5
    
class group:
    def __init__(self, group_name):
        self.group_name = group_name
        self.teamlist_group = []
    def sim_match(self, team1, team2):    #function to simulate match between 2 teams
        dummy = 0
        while dummy == 0:
            winner = input(f"{self.group_name} Match: {team1.country} vs. {team2.country} winner: ")
            winner = winner.upper()    # makes whatever user typed not case-sensitive
            if winner == "NONE" or winner == "NON":
                team1.upd_res("draw")
                team2.upd_res("draw")
                dummy = 1
            elif winner == team1.country.upper() or winner == team1.initials:
                team1.upd_res("win")
                team2.upd_res("loss")
                dummy = 1
            elif winner == team2.country.upper() or winner == team2.initials:
                team2.upd_res("win")
                team1.upd_res("loss")
                dummy = 1
            else:
                print("Please enter a valid input.")    #prompts the user again if their input isn't accepted
                
class round16:
    def __init__(self, round16):
        self.round16 = round16
        self.teamlist_r16 = []
    #function to enter a group's top 2 teams into the round of 16 stage
    def enter_teams_fromgroup(self, qualifiers_group: list[team]):
        score = {}
        for team in qualifiers_group:
            score[team] = team.win + team.draw - team.loss    #tallies win/draw/loss results of all teams in the group
        score_dict = dict(sorted(score.items(), key=lambda item: item[1], reverse=True))    #sorts teams by score
        score = list(score_dict.values())
        score_keys = list(score_dict.keys())
        #iterates through the 2 teams with the highest scores in the group
        for m in range(0, len(score_keys)-2):
            dummy = 0
            current_key = score_keys[m]
            if m != len(score_keys):
                next_key = score_keys[m+1]
            if score[m] > score[m+1]:
                self.teamlist_r16.append(score_keys[m])
            elif score[m] < score[m+1]:
                self.teamlist_r16.append(score_keys[m+1])
            elif score[m] == score[m+1]:
                if current_key.win == next_key.win and current_key.draw == next_key.draw:
                    #ask user which team scored more goals
                    while dummy == 0:
                        winner = input(f"{score_keys[m].country} ties {score_keys[m+1].country} \
for #{m+1} place. Team with better goal difference: ")
                        winner = winner.upper()
                        if winner == score_keys[m].country.upper() or winner == score_keys[m].initials:
                            self.teamlist_r16.append(score_keys[m])
                            dummy = 1
                        elif winner == score_keys[m+1].country.upper() or winner == score_keys[m+1].initials:
                            self.teamlist_r16.append(score_keys[m+1])
                            dummy = 1
                        else:
                            print("Please enter a valid input.")
                elif current_key.win > next_key.win or \
                current_key.win == next_key.win and current_key.draw > next_key.draw:
                    self.teamlist_r16.append(score_keys[m])
                elif current_key.win < next_key.win or \
                current_key.win == next_key.win and current_key.draw < next_key.draw:
                    self.teamlist_r16.append(score_keys[m+1])
    #similar function to sim_match() from before, but doesn't accept a draw result
    def sim_r16match(self, team1, team2, qf):
        dummy = 0
        while dummy == 0:
            winner = input(f"{self.round16} Match: {team1.country} vs. {team2.country} winner: ")
            winner = winner.upper()
            if winner == team1.country.upper() or winner == team1.initials:
                qf.enter_team_fromr16(team1)
                dummy = 1
            elif winner == team2.country.upper() or winner == team2.initials:
                qf.enter_team_fromr16(team2)
                dummy = 1
            else:
                print("Please enter a valid input.")

#classes for quarter-finals, semi-finals, and finals have similar structures to the round16 class
class qfinal:
    def __init__(self, qfinal):
        self.qfinal = qfinal
        self.teamlist_qf = []
    def enter_team_fromr16(self, team_toqf):
        self.teamlist_qf.append(team_toqf)
    def sim_qfmatch(self, team1, team2, sf):
        dummy = 0
        while dummy == 0:
            winner = input(f"{self.qfinal} Match: {team1.country} vs. {team2.country} winner: ")
            winner = winner.upper()
            if winner == team1.country.upper() or winner == team1.initials:
                sf.enter_teams_fromqf(team1)
                dummy = 1
            elif winner == team2.country.upper() or winner == team2.initials:
                sf.enter_teams_fromqf(team2)
                dummy = 1
            else:
                print("Please enter a valid input.")
                
class sfinal:
    def __init__(self, sfinal):
        self.sfinal = sfinal
        self.teamlist_sf = []
    def enter_teams_fromqf(self, team_tosf):
        self.teamlist_sf.append(team_tosf)
    def sim_sfmatch(self, team1, team2, fn):
        dummy = 0
        while dummy == 0:
            winner = input(f"{self.sfinal} Match: {team1.country} vs. {team2.country} winner: ")
            winner = winner.upper()
            if winner == team1.country.upper() or winner == team1.initials:
                fn.enter_teams_fromsf(team1)
                dummy = 1
            elif winner == team2.country.upper() or winner == team2.initials:
                fn.enter_teams_fromsf(team2)
                dummy = 1
            else:
                print("Please enter a valid input.")
                
class final:
    def __init__(self, final):
        self.final = final
        self.teamlist_fn = []
    def enter_teams_fromsf(self, team_tof):
        self.teamlist_fn.append(team_tof)
    def sim_fnmatch(self, team1, team2):
        dummy = 0
        while dummy == 0:
            winner = input(f"{self.final} Match: {team1.country} vs. {team2.country} winner: ")
            winner = winner.upper()
            if winner == team1.country.upper() or winner == team1.initials:
                print(f"The winner of the World Cup is {team1.country}!")
                dummy = 1
            elif winner == team2.country.upper() or winner == team2.initials:
                print(f"The winner of the World Cup is {team2.country}!")
                dummy = 1
            else:
                print("Please enter a valid input.")
    
#### Defines objects for teams, groups, and each stage
net = team("Netherlands", "NET")
sen = team("Senegal", "SEN")
ecu = team("Ecuador", "ECU")
qat = team("Qatar", "QAT")
eng = team("England", "ENG")
usa = team("United States", "USA")
irn = team("Iran", "IRN")
wal = team("Wales", "WAL")
arg = team("Argentina", "ARG")
pol = team("Poland", "POL")
mex = team("Mexico", "MEX")
ksa = team("Saudi Arabia", "KSA")
fra = team("France", "FRA")
aus = team("Australia", "AUS")
tun = team("Tunisia", "TUN")
den = team("Denmark", "DEN")
jpn = team("Japan", "JPN")
esp = team("Spain", "ESP")
ger = team("Germany", "GER")
crc = team("Costa Rica", "CRC")
mar = team("Morocco", "MAR")
cro = team("Croatia", "CRO")
bel = team("Belgium", "BEL")
can = team("Canada", "CAN")
bra = team("Brazil", "BRA")
sui = team("Switzerland", "SUI")
cmr = team("Cameroon", "CMR")
srb = team("Serbia", "SRB")
por = team("Portugal", "POR")
kor = team("South Korea", "KOR")
uru = team("Uruguay", "URU")
gha = team("Ghana", "GHA")
a = group("Group A")
a.teamlist_group = [net, sen, ecu, qat]
b = group("Group B")
b.teamlist_group = [eng, usa, irn, wal]
c = group("Group C")
c.teamlist_group = [arg, pol, mex, ksa]
d = group("Group D")
d.teamlist_group = [fra, aus, tun, den]
e = group("Group E")
e.teamlist_group = [jpn, esp, ger, crc]
f = group("Group F")
f.teamlist_group = [mar, cro, bel, can]
g = group("Group G")
g.teamlist_group = [bra, sui, cmr, srb]
h = group("Group H")
h.teamlist_group = [por, kor, uru, gha]
r16 = round16("Round of 16")
qf = qfinal("Quarter-Finals")
sf = sfinal("Semi-Finals")
fn = final("Finals")

### Simulates each round's matches
print("NOTE: Country initials can be used as well, and input isn't case-sensitive. Type \'None\' if match is a draw.")

#groupmatches
#creates function to simulate all possible matches in a group and enters top 2 teams into the round of 16 stage
def groupmatch(i, n):
    for j in itertools.combinations(i.teamlist_group, 2):
        k = list(j)
        i.sim_match(k[0], k[1])
    r16.enter_teams_fromgroup(i.teamlist_group)
    print(f"Qualifiers from {i.group_name}: {r16.teamlist_r16[n*2-2].country}, {r16.teamlist_r16[n*2-1].country}")

groups = [a, b, c, d, e, f, g, h]
w = 0
for group in groups:
    groupmatch(group, w)
    w += 1

#rounds of 16
x = 0
#iterates through all possible matches in the round of 16 stage
while x < 16:
    if x % 2 != 0: #odd number
        r16.sim_r16match(r16.teamlist_r16[x], r16.teamlist_r16[x+1], qf)
        x = x + 3
    else: #even number
        r16.sim_r16match(r16.teamlist_r16[x], r16.teamlist_r16[x+3], qf)
        x += 1
print(f"Quarter-Finals qualifiers: {qf.teamlist_qf[0].country}, {qf.teamlist_qf[1].country}, \
{qf.teamlist_qf[2].country}, {qf.teamlist_qf[3].country}, {qf.teamlist_qf[4].country}, \
{qf.teamlist_qf[5].country}, {qf.teamlist_qf[6].country}, {qf.teamlist_qf[7].country}")

#quarterfinals
y = 0
while y < 8:
    if y % 2 != 0: #odd number
        qf.sim_qfmatch(qf.teamlist_qf[y], qf.teamlist_qf[y+2], sf)
        y = y + 3
    else: #even number
        qf.sim_qfmatch(qf.teamlist_qf[y], qf.teamlist_qf[y+2], sf)
        y += 1
print(f"Semi-Finals qualifiers: {sf.teamlist_sf[0].country}, {sf.teamlist_sf[1].country}, \
{sf.teamlist_sf[2].country}, {sf.teamlist_sf[3].country}")

#semifinals
z = 0
while z < 4:
    if z % 2 != 0: #odd number
        sf.sim_sfmatch(sf.teamlist_sf[z], sf.teamlist_sf[z+2], fn)
        z = z + 3
    else: #even number
        sf.sim_sfmatch(sf.teamlist_sf[z], sf.teamlist_sf[z+2], fn)
        z += 1
print(f"Finals qualifiers: {fn.teamlist_fn[0].country}, {fn.teamlist_fn[1].country}")

#finals
fn.sim_fnmatch(fn.teamlist_fn[0], fn.teamlist_fn[1])
```
